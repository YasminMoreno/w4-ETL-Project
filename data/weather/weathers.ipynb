```python
import requests as req

import pandas as pd

import numpy as np

from bs4 import BeautifulSoup
```

```python
canada=pd.read_excel('../weather/canada_clima.xls')
usa=pd.read_excel('../weather/usa_clima.xls')
argentina=pd.read_excel('../weather/argentina_clima.xls')
colombia=pd.read_excel('../weather/colombia_clima.xls')
spain=pd.read_excel('../weather/spain_clima.xls')
uk=pd.read_excel('../weather/uk_clima.xls')
alemania=pd.read_excel('../weather/alemania_clima.xls')
australia=pd.read_excel('../weather/australia_clima.xls')
korea=pd.read_excel('../weather/korea_clima.xls')
japon=pd.read_excel('../weather/japon_clima.xls')


```

```python
weather_1 = pd.concat([canada, usa, argentina, colombia, spain, uk, alemania, australia, korea, japon], axis=0)

weather_1
```

```python
weather = weather_1.drop(['Hour', 'Borrar', 'Precip', 'Dew Point'], axis =1)

weather
```

```python
weather['id_weather']=np.arange(1,len(weather)+1)
weather
```

```python
first_column = weather.pop('id_weather') 
  
weather.insert(0, 'id_weather', first_column)
weather
```

```python
weather.to_excel('weather.xls', index=False)
```
