```python
df_pandas=pd.read_clipboard()

df_pandas
```

```python
countries = df_pandas.drop(['Sr. No.'], axis = 1)
countries
```

```python
countries = countries.rename({'Country Name': 'country', 'Capital': 'capital', 'Currency': 'currency'},  axis=1)
countries
```

```python
countries['id_country']=np.arange(1,len(countries)+1)
countries
```

```python
first_column = countries.pop('id_country') 
  
countries.insert(0, 'id_country', first_column)
countries
```

```python
countries.to_excel('countries.xls', index=False)
```
